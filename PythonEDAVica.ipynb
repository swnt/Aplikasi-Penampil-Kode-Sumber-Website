```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
```

## Connect to Datasource
- url = ' ' import dataset
- df -> variabel dataframe
- pd -> panggil pandas library
- read_csv() -> membaca file dalam format csv

```python
url = 'https://raw.githubusercontent.com/KennethParulian/PythonEDAEnigmacamp/main/MOCK_DATA%20(4).csv'

df = pd.read_csv(url)
```

```python
df.head(20)
```

```python
print(df.isnull().sum())
```

- Drop data null

```python
df.dropna(inplace=True)
df.count()
```

```python
df.rename({'id': 'Id', 
           'first_name': 'Nama Depan', 
           'last_name': 'Nama Belakang', 
           'gender': 'Gender',
           'CreditCardNumber': 'Credit Card Number',
           'CreditCardLimit': 'Credit Card Limit',
           }, axis=1, inplace=True)
df.head()
```

- menambah column
- np.random -> untuk membuat string/int random
- randint -> type random int
- (1, 10) -> untuk range angka randomnya
- df.shape[0/1] -> untuk menunjukan tabel horizontal atau vertical

```python
df['Credit Rating'] = np.random.randint(1, 10, df.shape[0])
df.head()
```

```python
df['Credit Remaining'] = np.random.randint(1000, 15000, df.shape[0])
df.head()
```

```python
df['Credit Card Limit'] = df['Credit Card Limit'].replace('[\$\,\.]', "", regex=True).astype(int)
df
```

```python
gender = df.groupby(['Gender'])['Id'].count().sort_values(ascending=True)

plt.barh(gender.index, gender)
plt.grid(linestyle='--', linewidth=2, axis='y',alpha = 0.3)
plt.ylabel("Gender")
plt.show()
```

```python
maxMinLimitGender = df.groupby('Gender').agg({'Credit Card Limit': ['min', 'max']}).plot.barh();
```

```python
meanLimitGender = df.groupby('Gender').agg({'Credit Card Limit': ['mean']}).plot.barh();
```
